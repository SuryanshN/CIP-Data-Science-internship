# independent and dependent features

```python

x=df_train.iloc[:,:-1]
y=df_train['price_range']
```

```python
test_data=df_test.drop('id',axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
test_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
St=StandardScaler()
x_train_scaled=St.fit_transform(x)
```

```python
x_test_scaled=St.transform(test_data)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train_scaled,y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# model training
```

```python
LR=LogisticRegression()
```

```python
LR.fit(x_train_scaled,y)
```

```python
# train with given test data
pred=LR.predict(x_test_scaled)
```

```python
# now train the model with data we have.
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
y_pred
```

```python
conf_matrix=confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
report=classification_report(y_test,y_pred)
```

```python
report
```
